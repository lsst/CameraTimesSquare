```python
repo_path = "/repo/embargo"
instrument="LSSTCam"
program='BLOCK-T654'
fields = "science_program,group,day_obs,seq_num"
```

```python
from lsst.daf.butler import Butler
from astropy.table import Table
```

```python
fields=fields.split(",")
butler = Butler(repo_path,collections=["LATISS/raw/all","LSSTCam/raw/all"])
registry = butler.registry
where=f"instrument='{instrument}' AND science_program='{program}'"
refs = list(
    butler.registry.queryDimensionRecords(
        'exposure',
        where=where
    )
)
```

```python
table = Table([ ref.toDict() for ref in refs ])
table_grouped = table.group_by('group')
```

```python
for key, group in zip(table_grouped.groups.keys, table_grouped.groups):
    print(f'****** {key["group"]} *******', group['seq_num'].min(), group['seq_num'].max())
    print(group)
    print('')
```
